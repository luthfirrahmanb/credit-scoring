## 9. Isi missing values Type 2 berdasarkan sum dari mean HP dan Speed terdekat dengan sum HP dan Speednya 

#### Show dataframe info and value_counts of Type 2 before filling missing values

```python
df_pokemon.info()
```

```python
df_pokemon['Type 2'].value_counts()
```

```python
sumSpeedAndHp = df_pokemon.groupby('Type 2')['Speed'].mean() + df_pokemon.groupby('Type 2')['HP'].mean()
sumSpeedAndHp
```

```python
df_pokemon['HP+Speed'] = df_pokemon['HP'] + df_pokemon['Speed']
```

```python
df_pokemon[df_pokemon['Type 2'].isna()].head()
```

```python
dictionarySums = list(sumSpeedAndHp.values)
dictionarySums = np.array(dictionarySums)
dictionarySums
```

```python
def find_nearest(value):
    array = dictionarySums
    types = ""
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    if idx == 0:
        types = "Bug"
    elif idx == 1:
        types = "Dark"
    elif idx == 2:
        types = "Dragon"
    elif idx == 3:
        types = "Electric"
    elif idx == 4:
        types = "Fairy"
    elif idx == 5:
        types = "Fighting"
    elif idx == 6:
        types = "Fire"
    elif idx == 7:
        types = "Flying"
    elif idx == 8:
        types = "Ghost"
    elif idx == 9:
        types = "Grass"
    elif idx == 10:
        types = "Ground"
    elif idx == 11:
        types = "Ice"
    elif idx == 12:
        types = "Normal"
    elif idx == 13:
        types = "Poison"
    elif idx == 14:
        types = "Psychic"
    elif idx == 15:
        types = "Rock"
    elif idx == 16:
        types = "Steel"
    elif idx == 17:
        types = "Water"
    return types
```

```python
nearestValues = df_pokemon[df_pokemon['Type 2'].isna()]['HP+Speed'].apply(find_nearest)
```

```python
df_pokemon['Type 2'].fillna(nearestValues, inplace=True)
```

```python
# nanage = dataset.groupby('who')['age'].mean()
# dataset['age'].fillna(dataset['who'].map(round(nanage)),inplace=True)
df_pokemon.drop('HP+Speed', axis=1, inplace=True)
```

## 10. Do MinMax Scaler for all continous variable except (#)

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df_pokemon[df_pokemon.columns[4:11]])
scaled_features = scaler.transform(df_pokemon[df_pokemon.columns[4:11]])
dataScaled = pd.DataFrame(scaled_features,columns=df_pokemon.columns[4:11])
df_pokemon.drop(df_pokemon.columns[4:11], axis=1, inplace=True)
df_pokemon=pd.concat([dataScaled,df_pokemon], axis=1)
```

## 11. Do one hot encoding for Type 1 Feature and binary encoding for Type 2 Feature
## dont forget to drop first feature from binary encoding

```python
Type1 = pd.get_dummies(df_pokemon['Type 1'])
df_pokemon = pd.concat([df_pokemon, Type1],axis=1)
df_pokemon.drop('Type 1', axis=1, inplace=True)
df_pokemon.head()
```

## 12. Train Test Split (test_size=0.30, random_state=101)
#### data = All Features except #, Name, and Legendary
#### target = Legendary

```python
from sklearn.model_selection import train_test_split
data = df_pokemon.drop(['#', 'Name', 'Legendary'], axis=1)
target = df_pokemon['Legendary']

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size=0.30, random_state=101)
```

## 13. Show the value_counts of the y_train and do RandomOverSampling

```python
pd.Series(y_train).value_counts()
```

```python
from imblearn.over_sampling import RandomOverSampler
ros = RandomOverSampler(random_state=101)

X_ros, y_ros = ros.fit_sample(X_train, y_train)
```

```python
pd.Series(y_ros).value_counts()
```

## 14. Make XGBoostClassifier model (random_state=101,n_estimators=100,learning_rate=0.1)
### fit the model with X_ros and y_ros

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, matthews_corrcoef, roc_auc_score, confusion_matrix, accuracy_score

from xgboost import XGBClassifier
xgb = XGBClassifier(random_state=101,n_estimators=100,learning_rate=0.1)
xgb.fit(X_ros, y_ros)
```

## 15. Try Predict and print the Confusion Matrix, Classification report, ROC AUC Score, Accuracy Score, and Matthews corrcoef

```python
predict = xgb.predict(X_ros)
predictProb = xgb.predict_proba(X_ros)
```

```python
print(confusion_matrix(y_ros,predict))
print(classification_report(y_ros, predict))
print(f"Roc Auc Score: {roc_auc_score(y_ros, predictProb[:,1])}")
print(f"Accuracy score: {accuracy_score(y_ros, predict)}")
print(f"Matthew corrcoef score: {matthews_corrcoef(y_ros, predict)}")
```
